```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

from ipywidgets import interact, IntSlider, fixed, FloatSlider

from scipy import ndimage, misc
import tkinter
from PIL import Image, ImageDraw, ImageTk
```

```python
def resizeCalc(size, maxSize = 300):
    ratio = min(maxSize/size[0],maxSize/size[1])
    return (np.int(size[0]*ratio),np.int(size[1]*ratio))

# Создаем объект для кластеризации.
#db=DBSCAN(eps=4, min_samples=3, n_jobs=4)
db=KMeans(n_clusters=100)

def clusterize(pixInitial):
    data = [(x, y, pixInitial[x, y][0], pixInitial[x, y][1], pixInitial[x, y][2]) for x in range(pixInitial.shape[0]) for y in range(pixInitial.shape[1])]
    db_clusters=db.fit_predict(data)
    return data, db_clusters
```

```python
selectedPoint = (0,0)
selectedColor = (199, 89, 28)


image = Image.open('data2/orange1.jpg')
image = image.resize(resizeCalc(image.size))

pixInitial = image.load()
image
```

```python
%%time

data, db_clusters = clusterize(np.array(image))
```

```python
for p, c in zip(data, db_clusters):
    pixInitial[p[0], p[1]] = (int(db.cluster_centers_[c, 2]), 
                        int(db.cluster_centers_[c, 3]), 
                        int(db.cluster_centers_[c, 4]))
    
plt.rcParams['figure.figsize'] = (12, 8)    
ax = plt.subplot()
ax.imshow(image);
```

```python

def selectColor(image, pixelAccess):
    global selectedPoint, selectedColor

    root = tkinter.Tk()

    # создаем рабочую область
    frame = tkinter.Frame(root)
    frame.grid()

    #Добавим изображение
    canvas = tkinter.Canvas(root, height=image.size[0], width=image.size[1])
    photo = ImageTk.PhotoImage(image)
    imageCv = canvas.create_image(0,0, anchor='nw',image=photo)
    canvas.grid(row=1,column=0,columnspan=2)
    
    lbl = tkinter.Label(root, text="Point: " + str(selectedPoint[0]) + ", " + str(selectedPoint[1]))  
    lbl.grid(column=0, row=0,sticky='w') 
    
    color = tkinter.Canvas(root, height=30, width=np.int(image.size[1]/2),background='#%02x%02x%02x' % selectedColor)
    color.grid(column=1,row=0,sticky='e')
    
    btn = tkinter.Button(root, text="OK", command=root.destroy)
    btn.grid(column=1,row=2,sticky='e')
    
    def click(event):
        global selectedPoint, selectedColor
        selectedPoint = (event.x, event.y)
        selectedColor = pixelAccess[selectedPoint]

        lbl.configure(text="Point: " + str(selectedPoint[0]) + ", " + str(selectedPoint[1]))
        color.configure(background='#%02x%02x%02x' % selectedColor)

    canvas.bind('<1>',click)   

    root.mainloop()

    selectedColor = pixelAccess[selectedPoint]
    
    return selectedColor

selectedColor = selectColor(image,pixInitial)
# selectedColor = (199, 89, 28)

print("Selected point: ", selectedPoint)
print("Selected color: ",selectedColor)
```

```python
rangeList = list(range(1,7))
rangeSize = len(rangeList)


colCount = 2
rowCount = np.int(np.ceil(rangeSize/colCount))

print(rangeList,rangeSize)
print(colCount,rowCount)

fig, axes = plt.subplots(rowCount,colCount,figsize=(24,12))
pixes = np.zeros(rangeSize,dtype=object)

for i,n in zip(range(rangeSize), rangeList):
    image2 =  Image.open('data2/orange'+ str(n) + '.jpg')
    resizedImageArray = np.array(image2.resize(resizeCalc(image2.size)))
    pixes[i] = resizedImageArray
    axes[np.int(i/colCount)][i%colCount].imshow(pixes[i]);

```

```python
%%time
pixesData = np.zeros(rangeSize,dtype=object)
pixesClusters = np.zeros(rangeSize,dtype=object)
pixesResults = np.zeros(rangeSize,dtype=object)

threshold = 90
figResult, axesResults = plt.subplots(rowCount,colCount,figsize=(24,12))

for k, pix in zip(range(rangeSize), pixes):
    print(k,' clusterize')
    pixesData[k], pixesClusters[k] = clusterize(pix)
    
    print(k,' find near')
    near = []
    for i, c in enumerate(db.cluster_centers_):
#         if abs(c[2]-selectedColor[0])+abs(c[3]-selectedColor[1])+abs(c[4]-selectedColor[2])<threshold:
        if selectedColor[0] - threshold < c[2] < selectedColor[0] + threshold and selectedColor[1] - threshold < c[3] < selectedColor[1] + threshold and selectedColor[2] - threshold < c[4] < selectedColor[2] + threshold:
#             print(i, c)
            near.append(i)

    print(k,' draw')
    pixesResults[k] = np.zeros(pix.shape, np.uint8)

    for p, c in zip(pixesData[k], pixesClusters[k]):
        if c in near:
            pixesResults[k][p[0], p[1]] = pix[p[0], p[1]]
        else:
            av = np.int((pix[p[0], p[1]][0]+pix[p[0], p[1]][1]+pix[p[0], p[1]][2]) / 3)
            pixesResults[k][p[0], p[1]] = (av,av,av)
    
    print(k,' show')
    axesResults[np.int(k/colCount)][k%colCount].imshow(pixesResults[k]);
```

```python
figCompare, axesCompare = plt.subplots(rangeSize,2,figsize=(20,30))
for k in range(rangeSize):
    print(k)
    axesCompare[k][0].imshow(pixes[k])
    axesCompare[k][1].imshow(pixesResults[k])
```
