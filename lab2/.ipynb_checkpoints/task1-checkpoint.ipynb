```python
try:    
    cap = cv2.VideoCapture(0)
    # params for ShiTomasi corner detection
    feature_params = dict( maxCorners = 100,
                           qualityLevel = 0.1,
                           minDistance = 5,
                           blockSize = 7 )
    # Parameters for lucas kanade optical flow
    lk_params = dict( winSize  = (15,15),
                      maxLevel = 2,
                      criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03))
    # Create some random colors
    color = np.random.randint(0,255,(100,3))
    # Take first frame and find corners in it
    ret, old_frame = cap.read()  # penalt frame
    ret, old_frame = cap.read()
    prev_frame = old_frame
    
    old_gray = cv2.cvtColor(old_frame, cv2.COLOR_BGR2GRAY)
    p0 = cv2.goodFeaturesToTrack(old_gray, mask = None, **feature_params)
    # Create a mask image for drawing purposes
    mask = np.zeros_like(old_frame)

    # Будем считать кадры.
    cl = 0
    
    
    fin12,fin02,fin18,fin08 = (np.int(old_frame.shape[1]*0.2), np.int(old_frame.shape[0]*0.2),np.int(old_frame.shape[1]*0.8), np.int(old_frame.shape[0]*0.8))
    
    while(1):
        ret,frame = cap.read()
        frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        mask = np.zeros(frame.shape)
        gray2 =  np.zeros(frame.shape, np.uint8)
#         testFr =  np.zeros((abs(finA-finC),abs(finB-finD),3), np.uint8)
        testFr =  np.zeros(frame.shape, np.uint8)
        
        # calculate optical flow
        p1, st, err = cv2.calcOpticalFlowPyrLK(old_gray, frame_gray, p0, None, **lk_params)
        if(p1 is None):
            p0 = cv2.goodFeaturesToTrack(old_gray, mask = None, **feature_params)
            continue

        # Select good points
        good_new = p1[st==1]
        good_old = p0[st==1]
        # draw the tracks
        dist = np.zeros((good_new.shape[0],2))
        for i,(new,old) in enumerate(zip(good_new,good_old)):
            a,b = new.ravel()
            c,d = old.ravel()
#             mask = cv2.line(mask, (a,b),(c,d), color[i].tolist(), 2)
            gray2 = cv2.circle(gray2,(a,b),5,color[i].tolist(),-1)
            dist[i] = [np.int(a-c),np.int(b-d)]
    #         dist[i] = ((a-c)**2 + (b-d)**2)**0.5

    #         text = str(dist[i][0]) + "; " + str(dist[i][1])
    #         cv2.putText(frame, text, (50,80 + i*30),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (200, 10, 20), 2)

        med0 = np.int(median(dist[:,0]))
        med1 = np.int(median(dist[:,1]))

        edge = 3

        testPX = (0,frame.shape[1])
        testPY = (0,frame.shape[0])
        
        if(med0 < -edge):
            pointA = (0, 0)
            pointB = (np.int(abs(med0)), frame.shape[0])
#             cv2.rectangle(frame, pointA, pointB, (20,30,200), -1)
            cv2.rectangle(mask, pointA, pointB, (255,255,255), -1)
    
            testPX = (np.int(abs(med0)), frame.shape[1])
            
        elif(med0 > edge):
            pointA = (np.int(frame.shape[1] - med0), 0)
            pointB = (frame.shape[1], frame.shape[1])
#             cv2.rectangle(frame, pointA, pointB, (20,30,200), -1)
            cv2.rectangle(mask, pointA, pointB, (255,255,255), -1)

            testPX = (0, np.int(frame.shape[1] - med0))
    
        if(med1 < -edge):
            pointC = (0, np.int(frame.shape[0]+med1 ))
            pointD = (frame.shape[1], frame.shape[0])
#             cv2.rectangle(frame, pointA, pointB, (20,30,200), -1)
            cv2.rectangle(mask, pointC, pointD, (255,255,255), -1)
            
            testPY = (0, np.int(frame.shape[0]+med1 ))
            
        elif(med1 > edge):
            pointC = (0, 0)
            pointD = (frame.shape[1], np.int(med1))
#             cv2.rectangle(frame, pointA, pointB, (20,30,200), -1)
            cv2.rectangle(mask, pointC, pointD, (255,255,255), -1)
    
            testPY =  (np.int(med1),frame.shape[0])
            
        cv2.rectangle(frame, (np.int(mask.shape[1]*0.2 + med0) ,np.int(mask.shape[0]*0.2 + med1)),(np.int(mask.shape[1]*0.8 + med0) ,np.int(mask.shape[0]*0.8 + med1)),(0,255,255))
        
        if(fin02+med1>fin08+med1):
            startX = fin08+med1
            endX = fin02+med1
        else:
            startX = fin02+med1
            endX = fin08+med1

        if(fin12+med0>fin18+med0):
            startY = fin18+med0
            endY = fin12+med0
        else:
            startY = fin12+med0
            endY = fin18+med0

        cropped = frame[startX:endX,startY:endY]
        

        cl = (cl + 1)
        cv2.putText(frame, str(cl), (50, 25),cv2.FONT_HERSHEY_SIMPLEX, 1, (200, 10, 20), 2)
        cv2.putText(frame, "median: " + str(med0) + ", " + str(med1), (100, 25),cv2.FONT_HERSHEY_SIMPLEX, 1, (200, 10, 200), 2)

        gray2 = np.where(mask==0, frame,gray2)
        gray2 = np.where(mask==255, prev_frame,gray2)
        
        mask = np.asarray(mask, np.uint8)
        img = cv2.add(frame,mask)
        
        if(cropped.shape[0]>0 and cropped.shape[1]>0):
            cv2.imshow('crop',cropped)
        cv2.imshow('frame',img)
#         cv2.imshow('gra',gray2)
#         cv2.imshow('mask',mask)
#         cv2.imshow('old',old_gray)
        k = cv2.waitKey(30) & 0xff
        if k == 27:
            break
        # Now update the previous frame and previous points

        if(cl > 100):
            cl = 0
            old_gray = frame_gray.copy()
            p0 = good_new.reshape(-1,1,2)
#             p0 = cv2.goodFeaturesToTrack(old_gray, mask = None, **feature_params)

        if(p0.shape[0] < 10):
            p0 = cv2.goodFeaturesToTrack(old_gray, mask = None, **feature_params)
        prev_frame = frame
except ArithmeticError:
    print("error")
finally:
    cv2.destroyAllWindows()
    cap.release()
    
```

```python
detector = cv2.SIFT()
norm = cv2.NORM_L2
matcher = cv2.BFMatcher(norm)

def filter_matches(kp1, kp2, matches, ratio = 0.75):
    mkp1, mkp2 = [], []
    for m in matches:
        if len(m) == 2 and m[0].distance < m[1].distance * ratio:
            m = m[0]
            mkp1.append( kp1[m.queryIdx] )
            mkp2.append( kp2[m.trainIdx] )
    p1 = np.float32([kp.pt for kp in mkp1])
    p2 = np.float32([kp.pt for kp in mkp2])
    kp_pairs = zip(mkp1, mkp2)
    return p1, p2, kp_pairs

def detectandselectmatches(fr1a,fr2a):
    kp1, desc1 = detector.detectAndCompute(fr1a, None)
    kp2, desc2 = detector.detectAndCompute(fr2a, None)
    raw_matches = matcher.knnMatch(desc1, trainDescriptors = desc2, k = 2) #2
    print("live")
    p1, p2, kp_pairs = filter_matches(kp1, kp2, raw_matches)
    return p1, p2

qx,qy=4,4
k=np.ones((qx,qy))

def preprocess(im1):
    im1 = cv2.GaussianBlur(im1, (25, 25), 0)
    return cv2.cvtColor(im1,cv2.COLOR_BGR2GRAY)
#     return cv2.erode(im1g,k)


def detectshift(im1,im2):
    p1, p2 = detectandselectmatches(im1,im2)
    dp=p2-p1
    if len(dp)<=0: return 0,0
    mx,my=np.median(dp[:,0]),np.median(dp[:,1])
    return mx,my

def getshiftmatrix(dshift):
    dx,dy = dshift
    return array([[ 1.,  0,   dx],
                  [ 0,  1.,   dy]])


def shiftimg(im2,shift):
    tr1=getshiftmatrix(shift)
    return cv2.warpAffine(im2,tr1,tuple(reversed(im2.shape[:2]) ))
```

```python
cap = cv2.VideoCapture(0)

img2 = cv2.imread("./rabit.png", -1)
img2 = cv2.resize(img2,(200,200),interpolation = cv2.INTER_CUBIC)
cv2.imshow('rab',img2)
alpha = img2[:, :, 3] / 255.0

# take first frame of the video
while(1):
    ret,frame = cap.read()
    if(ret != True):
        continue
    
    med0 = 200
    med1 = 0

    edge = 3

    if(med0 < -edge):
        pointA = (0, 0)
        pointB = (np.int(abs(med0)), frame.shape[0])
        cv2.rectangle(frame, pointA, pointB, (20,30,200), -1)

    elif(med0 > edge):
        pointA = (np.int(frame.shape[1] - med0), 0)
        pointB = (frame.shape[1], frame.shape[1])
        cv2.rectangle(frame, pointA, pointB, (20,30,200), -1)

    if(med1 < -edge):
        pointA = (0, np.int(frame.shape[0]+med1 ))
        pointB = (frame.shape[1], frame.shape[0])
        cv2.rectangle(frame, pointA, pointB, (20,30,200), -1)
    elif(med1 > edge):
        pointA = (0, 0)
        pointB = (frame.shape[1], np.int(med1))
        cv2.rectangle(frame, pointA, pointB, (20,30,200), -1)
    
    frame[0:200, 0:200, 0] = (1. - alpha) * frame[0:200, 0:200, 0] + alpha * img2[:, :, 0]
    frame[0:200, 0:200, 1] = (1. - alpha) * frame[0:200, 0:200, 1] + alpha * img2[:, :, 1]
    frame[0:200, 0:200, 2] = (1. - alpha) * frame[0:200, 0:200, 2] + alpha * img2[:, :, 2]
    
    cv2.imshow('frame',frame)
    
    
    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
